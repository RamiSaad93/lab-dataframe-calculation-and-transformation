## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
orders.columns
```

```python
# Sub Problem 1: How to aggregate the amount_spent for unique customers?

df_total = pd.DataFrame(orders.groupby(['CustomerID', 'Country'])['amount_spent'].sum().sort_values(ascending = False).reset_index())
df_total.columns = ['CustomerID','Country','total_spent']
df_total.head(60)
```

```python
# Sub Problem 2: How to select customers whose aggregated amount_spent is in a given quantile range?

# VIP are the top 10 spenders
# Preferred are the top 5% spenders without the VIP

vip_customers = df_total.head(10)
vip_customers

non_vip_customers = df_total.drop(index = range(10)).reset_index(drop = True)

preferred_customers = non_vip_customers[non_vip_customers['total_spent'] > df_total['total_spent'].quantile(0.95)]
preferred_customers
```

```python
preferred_customers_list = preferred_customers['CustomerID'].tolist()
vip_customers_list = vip_customers['CustomerID'].tolist()
```

```python
# Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

df_total['Customer_Type'] = df_total['CustomerID'].map(lambda x: 'Preferred' if x in preferred_customers_list
                                                      else ('VIP' if x in vip_customers_list else 'Ordinary'))
df_total.head(60)
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
vip_customers.groupby('Country')['CustomerID'].count().sort_values(ascending = False)
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
df_total[df_total['Customer_Type'].isin(['VIP', 'Preferred'])].groupby('Country')['CustomerID'].count().sort_values(ascending = False)
```
